## 3. Fonction d'inspection des données

Plutôt que de répéter les mêmes commandes pour chaque DataFrame, nous allons créer une **fonction réutilisable** qui génère un rapport complet.

### 🎯 Ce que notre fonction doit faire :
1. Afficher les informations générales (`.info()`)
2. Afficher les statistiques descriptives (`.describe()`)
3. Vérifier la présence de valeurs manquantes (NA)
4. Identifier une potentielle clé unique (identifiant)

## Fonction de base

````python
def inspector(df):
    
    """
    Génère un rapport d'inspection complet pour un DataFrame.
    
    Paramètres:
    -----------
    df : pandas.DataFrame
        Le DataFrame à inspecter
    
    Retourne:
    ---------
    None (affiche le rapport dans le terminal)
    """
    
    # Debut de l'inspection
    print("#####     DEBUT DE L'INSPECTION     #####")
    
    # Structure du DataFrame avec df.shape
    print("\n----------------------")
    print("Structure du DataFrame:")
    print(f'Nb de lignes: {df.shape[0]} ; Nb de colonnes: {df.shape[1]}')
    
    # Types de données avec df.dtypes
    print("\n----------------------")
    print("Types de données:")
    print(df.dtypes)
    
    # Présence d'une potentielle clé unique (identifiant) avec df.nunique()
    print("\n----------------------")
    print("Nombre de valeurs uniques (pour identifiant PK / FK):")
    print(df.nunique())
    
    # Présence de valeurs manquantes avec df.isnull().sum()
    print("\n----------------------")
    print("Valeurs manquantes:")
    print(df.isnull().sum())
    
    # Fin de l'inspection
    print("\n#####     FIN DE L'INSPECTION     #####")
````

## Fonction avancée

````python
def inspecter_dataframe(df, nom="DataFrame"):
    """
    Génère un rapport d'inspection complet pour un DataFrame.
    
    Paramètres:
    -----------
    df : pandas.DataFrame
        Le DataFrame à inspecter
    nom : str
        Le nom du DataFrame (pour l'affichage)
    
    Retourne:
    ---------
    None (affiche le rapport dans le terminal)
    """
    
    # Séparateur visuel
    separateur = "=" * 70
    
    # === EN-TÊTE ===
    print(f"\n{separateur}")
    print(f"📊 RAPPORT D'INSPECTION : {nom.upper()}")
    print(f"{separateur}\n")
    
    # === 1. DIMENSIONS ===
    print(f"📐 DIMENSIONS")
    print(f"   - Nombre de lignes   : {df.shape[0]:,}")
    print(f"   - Nombre de colonnes : {df.shape[1]}")
    print()
    
    # === 2. TYPES DE DONNÉES ===
    print(f"📋 COLONNES ET TYPES")
    print("-" * 40)
    for col in df.columns:
        print(f"   {col:25} → {df[col].dtype}")
    print()
    
    # === 3. VALEURS MANQUANTES ===
    print(f"❓ VALEURS MANQUANTES (NA)")
    print("-" * 40)
    na_counts = df.isna().sum()
    na_total = na_counts.sum()
    
    if na_total == 0:
        print("   ✅ Aucune valeur manquante !")
    else:
        print(f"   ⚠️  Total de valeurs manquantes : {na_total:,}")
        for col in df.columns:
            if na_counts[col] > 0:
                pct = (na_counts[col] / len(df)) * 100
                print(f"   - {col:25} : {na_counts[col]:,} ({pct:.1f}%)")
    print()
    
    # === 4. RECHERCHE DE CLÉ UNIQUE ===
    print(f"🔑 RECHERCHE DE CLÉ UNIQUE (IDENTIFIANT)")
    print("-" * 40)
    cle_trouvee = False
    
    for col in df.columns:
        nb_valeurs_uniques = df[col].nunique()
        nb_lignes = len(df)
        
        # Une colonne est une clé unique si chaque valeur est différente
        if nb_valeurs_uniques == nb_lignes:
            print(f"   ✅ '{col}' est une clé unique ({nb_valeurs_uniques:,} valeurs uniques)")
            cle_trouvee = True
        elif nb_valeurs_uniques == nb_lignes - df[col].isna().sum():
            # Cas où il y a des NA mais les valeurs non-NA sont uniques
            print(f"   ⚠️  '{col}' pourrait être une clé (unique hors NA)")
            cle_trouvee = True
    
    if not cle_trouvee:
        print("   ❌ Aucune colonne n'est une clé unique")
        print("   💡 Les colonnes avec le plus de valeurs uniques :")
        # Afficher les 3 colonnes avec le plus de valeurs uniques
        for col in df.columns[:3]:
            print(f"      - {col}: {df[col].nunique():,} valeurs uniques")
    print()
    
    # === 5. STATISTIQUES DESCRIPTIVES ===
    print(f"📈 STATISTIQUES DESCRIPTIVES")
    print("-" * 40)
    
    # Colonnes numériques
    cols_numeriques = df.select_dtypes(include=[np.number]).columns.tolist()
    if cols_numeriques:
        print("\n   Colonnes numériques :")
        display(df[cols_numeriques].describe().round(2))
    
    # Colonnes catégorielles (texte)
    cols_categorielles = df.select_dtypes(include=['object']).columns.tolist()
    if cols_categorielles:
        print("\n   Colonnes catégorielles :")
        display(df[cols_categorielles].describe())
    
    print(f"\n{separateur}")
    print(f"Fin du rapport pour {nom}")
    print(f"{separateur}\n")
````

### 💡 Comprendre la fonction

Notre fonction `inspecter_dataframe()` utilise plusieurs méthodes Pandas :

| Méthode | Ce qu'elle fait |
|---------|----------------|
| `df.shape` | Retourne (nb_lignes, nb_colonnes) |
| `df.dtypes` | Types de chaque colonne |
| `df.isna().sum()` | Compte les valeurs manquantes par colonne |
| `df.nunique()` | Nombre de valeurs uniques par colonne |
| `df.describe()` | Statistiques descriptives |

## 5. Jointure des fichiers

### 🎯 Objectif
Créer un DataFrame unifié contenant toutes les informations : transactions + détails produits + détails clients.

### 🔗 Identifier les clés de jointure

Pour joindre des tables, il faut identifier les **colonnes communes** :

````python
# Afficher les colonnes de chaque table
print("Colonnes CUSTOMERS    :", list(customers.columns))
print("Colonnes PRODUCTS     :", list(products.columns))
print("Colonnes TRANSACTIONS :", list(transactions.columns))
````

### 🔍 Analyse des relations

```
┌─────────────┐         ┌──────────────────┐         ┌─────────────┐
│  CUSTOMERS  │         │   TRANSACTIONS   │         │  PRODUCTS   │
├─────────────┤         ├──────────────────┤         ├─────────────┤
│ client_id   │◄───────►│ client_id        │         │ id_prod     │
│ sex         │         │ id_prod          │◄───────►│ price       │
│ birth       │         │ date             │         │ categ       │
│             │         │ session_id       │         │             │
└─────────────┘         └──────────────────┘         └─────────────┘
```

**Clés de jointure identifiées :**
- `transactions.client_id` ↔ `customers.client_id`
- `transactions.id_prod` ↔ `products.id_prod`

### 📚 Rappel : Les types de jointures

| Type | Description | Utilisation |
|------|-------------|-------------|
| `inner` | Garde uniquement les lignes présentes dans les 2 tables | Par défaut, le plus strict |
| `left` | Garde toutes les lignes de la table de gauche | Conserver toutes les transactions |
| `right` | Garde toutes les lignes de la table de droite | Rarement utilisé |
| `outer` | Garde toutes les lignes des 2 tables | Union complète |

### 5.1 Étape 1 : Joindre transactions et products

````python
# Vérifier les noms de colonnes pour la jointure
print("Dans transactions, la colonne produit s'appelle :", 
      [col for col in transactions.columns if 'prod' in col.lower()])
print("Dans products, la colonne produit s'appelle :", 
      [col for col in products.columns if 'prod' in col.lower() or 'id' in col.lower()])
````

````python
# Jointure transactions + products
# On utilise 'left' pour garder toutes les transactions
# même si un produit n'est pas trouvé dans la table products

df_transaction_produit = transactions.merge(
    products,
    left_on='id_prod',      # Colonne dans transactions
    right_on='id_prod',     # Colonne dans products
    how='left'              # Type de jointure
)

print(f"✅ Jointure transactions + products effectuée")
print(f"   - Lignes avant : {len(transactions):,}")
print(f"   - Lignes après : {len(df_transaction_produit):,}")
print(f"   - Colonnes     : {list(df_transaction_produit.columns)}")
````

````python
# Vérifier s'il y a des produits non trouvés (NA dans price ou categ)
produits_manquants = df_transaction_produit['price'].isna().sum()
print(f"\n⚠️  Produits non trouvés dans la table products : {produits_manquants}")

if produits_manquants > 0:
    print("\nDétail des produits manquants :")
    display(df_transaction_produit[df_transaction_produit['price'].isna()]['id_prod'].value_counts().head(10))
````

### 5.2 Étape 2 : Joindre avec customers

````python
# Jointure avec customers
transactions_enrichies = df_transaction_produit.merge(
    customers,
    on='client_id',         # Colonne commune (même nom dans les 2 tables)
    how='left'
)

print(f"✅ Jointure avec customers effectuée")
print(f"   - Lignes : {len(transactions_enrichies):,}")
print(f"   - Colonnes : {list(transactions_enrichies.columns)}")
````

````python
# Vérifier s'il y a des clients non trouvés
clients_manquants = transactions_enrichies['sex'].isna().sum()
print(f"\n⚠️  Clients non trouvés dans la table customers : {clients_manquants}")
````
